```python
# orders - first dataframe
# users - second dataframe
# "how" is type of join. can be "left", "right" or "inner". maybe "outer" also :)
# "left_on" is "orders.user_id" field
# "right_on" is "user.id" field

df = orders.merge(users, how='left', left_on='user_id', right_on='id')
```

```python
# lists the columns that dataframe has

df.columns
```

```python
# "drop" remove columns from dataframe

drop_columns = ['id_x', 'birth_data']

df = df.drop(drop_columns, axis=1)
```
